```python
import pandas as pd
```

```python
data_flood = pd.read_csv('raw data/flood_2023_final.csv')
data_flood
```

```python
data_flood = data_flood[['Disaster Date','End Disaster Date','Province','District','Subdistrict']]
data_flood = data_flood.drop(index=0) 
data_flood = data_flood.reset_index(drop=True)
data_flood
```

```python
data_flood['Disaster Date'] = pd.to_datetime(data_flood['Disaster Date'], dayfirst=True, errors='coerce')
data_flood['End Disaster Date'] = pd.to_datetime(data_flood['End Disaster Date'], dayfirst=True, errors='coerce')


data_flood['duration_days'] = (data_flood['End Disaster Date'] - data_flood['Disaster Date']).dt.days
avg_duration = data_flood['duration_days'].mean(skipna=True)


data_flood['End Disaster Date'] = data_flood['End Disaster Date'].fillna(
    data_flood['Disaster Date'] + pd.to_timedelta(round(avg_duration), unit='D')
)


data_flood = data_flood.drop(columns=['duration_days'])

print("avg_duration:", avg_duration, "day")
data_flood
```

```python
north_provinces = [
    'เชียงราย', 'เชียงใหม่', 'น่าน', 'พะเยา', 'แพร่', 'แม่ฮ่องสอน', 'ลำปาง', 'ลำพูน',
    'อุตรดิตถ์'
]
data_flood = data_flood[data_flood['Province'].isin(north_provinces)]
data_flood
```

```python

# สร้าง list เก็บข้อมูลขยาย
expanded_rows = []

for _, row in data_flood.iterrows():
    date_range = pd.date_range(start=row['Disaster Date'], end=row['End Disaster Date'])
    for single_date in date_range:
        expanded_rows.append({
            'Date': single_date,
            'Province': row['Province'],
            'District': row['District'],
            'Subdistrict': row['Subdistrict'],
            'Label': 1
        })

# สร้าง DataFrame ใหม่
df_expanded = pd.DataFrame(expanded_rows)

# เรียงตามวันที่
df_expanded = df_expanded.sort_values(by='Date').reset_index(drop=True)
df_expanded
```

```python
data_wather = pd.read_csv('data/Climate_2023.csv')
print(data_wather['Province'].unique())
data_wather
```

```python
data_dams = pd.read_csv('data/Dams_2023.csv')
print(data_dams['Name of dam'].unique())
data_dams
```

```python
data_river = pd.read_csv('data/River_2023.csv')
print(data_river['Station'].unique())
data_river
```

```python
df_expanded['Date'] = pd.to_datetime(df_expanded['Date'], dayfirst=True)
data_river['Date'] = pd.to_datetime(data_river['Date'], dayfirst=False)

# เตรียมข้อมูลระดับน้ำ (df2_clean)
df2_clean = data_river[['Date', 'Station', 'Average of Water heights in meter']].copy()
df2_clean.rename(columns={'Average of Water heights in meter':'Water_Height'}, inplace=True)
df2_clean['Water_Height'] = pd.to_numeric(df2_clean['Water_Height'], errors='coerce')

# สร้าง pivot table จากวันที่กับ Station
df_wide = df2_clean.pivot_table(
    index='Date',
    columns='Station',
    values='Water_Height',
    aggfunc='mean'  # หรือ 'first' ถ้าไม่ต้องการเฉลี่ย
).reset_index()

# รวมกับ df_expanded โดย join แค่ Date
df_flood_river = pd.merge(df_expanded, df_wide, how='left', on='Date')
df_flood_river

```

```python
data_dams['Date'] = pd.to_datetime(data_dams['Date'], dayfirst=False)

df_dam_wide = data_dams.pivot_table(
    index='Date',
    columns='Name of dam',
    values='Water volume in dams (by %)',
    aggfunc='mean'
).reset_index()

# --- รวมข้อมูลเขื่อนเข้ากับ df_flood_river โดยใช้ Date ---
df_flood_river_dams = pd.merge(df_flood_river, df_dam_wide, how='left', on='Date')

# --- ดูข้อมูลผลลัพธ์ ---
df_flood_river_dams
```

```python
dataframe = pd.read_csv('data/dataframe_2023.csv')
dataframe
```

```python
dataframe['date'] = pd.to_datetime(dataframe['date'], dayfirst=True)
df_expanded['Date'] = pd.to_datetime(df_expanded['Date'], dayfirst=True)

# สร้าง key สำหรับ merge (ใช้หลายคอลัมน์)
merge_cols = ['date', 'Province', 'District', 'Subdistrict']

# เปลี่ยนชื่อคอลัมน์ Date ใน df_flood ให้ตรงกับ df_base
df_flood = df_expanded.rename(columns={'Date': 'date'})

# รวมข้อมูล โดยเอา label = max (ถ้ามี 1 จะได้ 1)
df_merged = pd.merge(
    dataframe, 
    df_flood[merge_cols + ['Label']], 
    on=merge_cols, 
    how='left'
)

# ถ้า label_flood เป็น NaN ให้แทนด้วย 0
df_merged['Label_y'] = df_merged['Label_y'].fillna(0)

# รวม Label: ถ้า base=0 และ flood=1 → ให้เป็น 1
df_merged['Label'] = df_merged[['Label_x', 'Label_y']].max(axis=1)

# ลบคอลัมน์ส่วนเกิน
df_final = df_merged.drop(columns=['Label_x', 'Label_y'])
df_final
```
