```python
import pandas as pd
```

```python
data_flood = pd.read_csv('raw data/flood_2023_final.csv')
data_flood
```

```python
data_flood = data_flood[['Disaster Date','End Disaster Date','Province','District','Subdistrict']]
data_flood = data_flood.drop(index=0) 
data_flood = data_flood.reset_index(drop=True)
data_flood
```

```python
data_flood['Disaster Date'] = pd.to_datetime(data_flood['Disaster Date'], dayfirst=True, errors='coerce')
data_flood['End Disaster Date'] = pd.to_datetime(data_flood['End Disaster Date'], dayfirst=True, errors='coerce')

# เติม End Disaster Date ที่เป็น NaT ด้วย 7 วันหลังจาก Disaster Date
data_flood['End Disaster Date'] = data_flood['End Disaster Date'].fillna(
    data_flood['Disaster Date'] + pd.Timedelta(days=7)
)

data_flood
```
